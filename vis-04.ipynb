### Шаг 11. Нарисуйте 2 гистограммы чаевых на обед и ланч. Расположите их рядом по горизонтали.

### Шаг 12. Нарисуйте 2 scatterplots (для мужчин и женщин), показав связь размера счета и чаевых, дополнительно разбив по курящим/некурящим. Расположите их по горизонтали.

```python
fig, axs = plt.subplots(ncols=2, figsize=(12,6)) # plot 4 graphs 

# simple scatter plot between two variables 
sns.scatterplot(x='total_bill', y='tip', data=tips[tips['sex'] == 'Male'], hue='smoker', ax=axs[1], )
axs[1].set_title('Male')

# group by time and show the groups with different colors
sns.scatterplot(x='total_bill', y='tip', data=tips[tips['sex'] == 'Female'], hue='smoker', ax=axs[0])
axs[0].set_title('Female')
```
